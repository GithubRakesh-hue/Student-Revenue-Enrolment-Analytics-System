```python
import pandas as pd
```

```python
#Load CSV File
students = pd.read_csv('students.csv')
students
```

```python
# Access Columns

students['name']
students[['name','course']]
students['fees']
students.dtypes[['roll','name','course','fees','city']]
```

```python
#  Access Rows
students.head(5)
students.tail(4)
students.iloc[3]
students.iloc[2:7]
```

```python
# Add New Column
students['gst']= students['fees'] * 0.18
students[['fees','gst']]
```

```python
# Calculate New Column
students['total_fees'] = students['fees'] + students['gst']
students[['fees','gst','total_fees']]
```

```python
# Add New Row
students.loc[len(students)] = [1,'Vaibhav','AWS',35000,'Pune',6400.0,41400.0]
students
```

```python
# Rename Columns
students=students.rename(columns={'name':'student_name','fees':'course_fee','city':'location'})
students
```

```python
# Filtering

# Single Condition
students[students['location']=='Pune']
students[students['course_fee']>30000]

# Multiple Conditions
students[(students['location'] == 'Pune') & (students['course_fee']>25000)]
students[(students['location'] == 'Mumbai') | (students['course']=='Java')]
students[(students['location'] == 'Nagpur') & (students['course']=='Data Science')]
```

```python
# Sorting
students.sort_values(by='course_fee')
students.sort_values(by='course_fee', ascending=False)
students.sort_values(by=['location','course_fee'])
```

```python
# Create Another CSV
#payments.csv
payments=pd.read_csv('payments.csv')
```

```python
# merge operation
student_data =pd.merge(students,payments,how='inner',on='roll')
pd.merge(students,payments,how='left',on='roll')
pd.merge(students,payments,how='right',on='roll')
pd.merge(students,payments,how='outer',on='roll')
```

```python
# Remaining Fees Calculation
remaing_fee =student_data['remaining_fees'] = student_data['course_fee'] - student_data['paid_fees']
student_data[['student_name','course_fee','paid_fees','remaining_fees']] 
```

```python
#  Complex Filtering

remaing_fee =student_data['remaining_fees'] = student_data['course_fee'] - student_data['paid_fees']
students[remaing_fee >5000]

students[(students['location'] == 'pune') & (remaing_fee >5000)]
student_data[(student_data['course'] == 'Python') & (student_data['paid_fees'] < 20000)]
#students[(students['location'] == 'Mumbai') | (students['location'] == 'Nagpur') & (students['course_fee'] > 30000)]
```
